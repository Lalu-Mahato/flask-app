```python
import psycopg2

try:
    conn = psycopg2.connect(
        host="localhost",
        database="testdb",
        user="postgres",
        password="postgres"
    )
    print('Connected to the database.')
except psycopg2.Error as e:
    print('Unable to connect to the database', e)
```

```python
def create_bank_branch_table():
    cursor = conn.cursor()

    query = """
        CREATE TABLE IF NOT EXISTS banks_master (
            id SERIAL PRIMARY KEY,
            branch_code INTEGER,
            branch_name VARCHAR(255) NOT NULL
        )
    """

    cursor.execute(query)
    conn.commit()
    cursor.close()
```

```python
def insert_bank_data(branch_code, branch_name):
    cursor = conn.cursor()
    query = "INSERT INTO banks_master(branch_code, branch_name) VALUES (%s, %s)"
    branchdata = (branch_code, branch_name)

    cursor.execute(query, branchdata)
    conn.commit()

    cursor.close()
```

```python
unique_branchs
```

```python
for (code, name) in unique_branchs.values:
    insert_bank_data(code, name)
```

```python
def create_prospect_table():
    cursor = conn.cursor()

    query = """
        CREATE TABLE IF NOT EXISTS prospects_master (
            id SERIAL PRIMARY KEY,
            loan_account_number BIGINT,
            name VARCHAR(255) NOT NULL,
            address VARCHAR(255) NOT NULL,
            mobile_number VARCHAR(255) NOT NULL
        )
    """

    cursor.execute(query)
    conn.commit()
    cursor.close()
```
